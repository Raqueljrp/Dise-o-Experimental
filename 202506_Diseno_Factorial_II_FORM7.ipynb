## **1. Diseno Factorial II**


Un **Diseño Factorial $2^k$** es un tipo de diseño experimental donde se estudian $k$ factores, cada uno con dos niveles (típicamente llamados "bajo" y "alto"). El objetivo es investigar los efectos principales de cada factor, así como las interacciones entre ellos, en una variable de respuesta.

En un diseño factorial $2^k$, se realizan experimentos en todas las $2^k$ combinaciones posibles de los niveles de los factores. Esto permite estimar los efectos de los factores y sus interacciones de manera eficiente.

Las principales ventajas de un diseño factorial $2^k$ incluyen:
* **Eficiencia:** Permite evaluar múltiples factores simultáneamente.
* **Información sobre interacciones:** Permite identificar si el efecto de un factor depende del nivel de otro factor.
* **Estimación de efectos:** Proporciona estimaciones de los efectos principales y de interacción.

Un **Diseño Factorial II** (o $2^2$) es un caso específico de un diseño factorial $2^k$ donde se estudian **dos** factores, cada uno con dos niveles. Las combinaciones experimentales son:
* Nivel bajo de A, Nivel bajo de B
* Nivel alto de A, Nivel bajo de B
* Nivel bajo de A, Nivel alto de B
* Nivel alto de A, Nivel alto de B

Estos diseños son útiles para identificar rápidamente los factores más importantes y sus interacciones en las primeras etapas de una investigación.

## **3. Modelo estadistico**

El modelo estadístico apropiado para un diseño factorial con tres factores (A, B, y C) con posibles diferentes números de réplicas ($n_{ijk}$) para cada combinación de tratamientos es:

$Y_{ijkl} = \mu + \alpha_i + \beta_j + \gamma_k + (\alpha\beta)_{ij} + (\alpha\gamma)_{ik} + (\beta\gamma)_{jk} + (\alpha\beta\gamma)_{ijk} + \epsilon_{ijkl}$

Donde:

*   $Y_{ijkl}$: Es la l-ésima observación de la variable de respuesta en el i-ésimo nivel del factor A, el j-ésimo nivel del factor B y el k-ésimo nivel del factor C.
*   $\mu$: Es la media general de la respuesta.
*   $\alpha_i$: Es el efecto del i-ésimo nivel del factor A ($i = 1, ..., a$, donde 'a' es el número de niveles de A).
*   $\beta_j$: Es el efecto del j-ésimo nivel del factor B ($j = 1, ..., b$, donde 'b' es el número de niveles de B).
*   $\gamma_k$: Es el efecto del k-ésimo nivel del factor C ($k = 1, ..., c$, donde 'c' es el número de niveles de C).
*   $(\alpha\beta)_{ij}$: Es el efecto de la interacción entre el i-ésimo nivel de A y el j-ésimo nivel de B.
*   $(\alpha\gamma)_{ik}$: Es el efecto de la interacción entre el i-ésimo nivel de A y el k-ésimo nivel de C.
*   $(\beta\gamma)_{jk}$: Es el efecto de la interacción entre el j-ésimo nivel de B y el k-ésimo nivel de C.
*   $(\alpha\beta\gamma)_{ijk}$: Es el efecto de la interacción de tres vías entre el i-ésimo nivel de A, el j-ésimo nivel de B y el k-ésimo nivel de C.
*   $\epsilon_{ijkl}$: Es el error aleatorio asociado a la l-ésima observación dentro de la combinación de tratamientos $(i, j, k)$, que se supone sigue una distribución normal con media cero y varianza constante ($\epsilon_{ijkl} \sim N(0, \sigma^2)$).

En tu caso específico, tienes un factor con 3 niveles (Factor A, $a=3$) y dos factores con 2 niveles (Factor B, $b=2$ y Factor C, $c=2$). La estructura de réplicas variará para cada combinación de niveles $(i, j, k)$, y esto se reflejará en el análisis, particularmente en el cálculo de las sumas de cuadrados si se realiza manualmente. Sin embargo, el modelo lineal general descrito arriba sigue siendo el adecuado.

## **4. Anova**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el dataframe con los datos del experimento
data = []

# Temperatura A: 100, 120, 140
# Presión B: 400, 450
# Tiempo C: 30, 35 minutos

# Para cada combinación de factores, agregamos las 3 réplicas
temperatures = [100, 120, 140]
pressures = [400, 450]
times = [30, 35]

# Datos según la tabla proporcionada
datos_organizados = {
    100: {
        400: {30: [34, 33, 33], 35: [27, 29, 28]},
        450: {30: [32, 32, 33], 35: [28, 28, 27]}
    },
    120: {
        400: {30: [32, 34, 34], 35: [26, 28, 29]},
        450: {30: [32, 33, 32], 35: [30, 25, 27]}
    },
    140: {
        400: {30: [36, 36, 37], 35: [29, 29, 30]},
        450: {30: [34, 34, 34], 35: [27, 28, 29]}
    }
}

# Construir el dataframe
for temp in temperatures:
    for pressure in pressures:
        for time in times:
            for rep, rendimiento in enumerate(datos_organizados[temp][pressure][time]):
                data.append({
                    'Temperatura': temp,
                    'Presion': pressure,
                    'Tiempo': time,
                    'Rendimiento': rendimiento,
                    'Replica': rep + 1
                })

df = pd.DataFrame(data)
print("DataFrame creado:")
print(df.head(10))
print(f"\nDimensiones del dataset: {df.shape}")

# Análisis exploratorio inicial
print("\n" + "="*50)
print("ANÁLISIS EXPLORATORIO")
print("="*50)

print(f"\nEstadísticas descriptivas:")
print(df.groupby(['Temperatura', 'Presion', 'Tiempo'])['Rendimiento'].describe())

# Gráficos exploratorios
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='Temperatura', y='Rendimiento', hue='Presion')
plt.title('Rendimiento por Temperatura y Presión')

plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='Tiempo', y='Rendimiento', hue='Temperatura')
plt.title('Rendimiento por Tiempo y Temperatura')

plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='Presion', y='Rendimiento', hue='Tiempo')
plt.title('Rendimiento por Presión y Tiempo')

plt.subplot(2, 2, 4)
# Gráfico de interacción triple
interaction_data = df.groupby(['Temperatura', 'Presion', 'Tiempo'])['Rendimiento'].mean().reset_index()
sns.lineplot(data=interaction_data, x='Temperatura', y='Rendimiento',
             hue='Presion', style='Tiempo', markers=True)
plt.title('Interacción entre Factores')

plt.tight_layout()
plt.show()

# ANOVA DE TRES VÍAS COMPLETO
print("\n" + "="*50)
print("ANOVA DE TRES VÍAS COMPLETO")
print("="*50)

# Convertir variables categóricas
df['Temperatura'] = df['Temperatura'].astype('category')
df['Presion'] = df['Presion'].astype('category')
df['Tiempo'] = df['Tiempo'].astype('category')

# Modelo ANOVA con todas las interacciones
model = ols('Rendimiento ~ C(Temperatura) + C(Presion) + C(Tiempo) + \
            C(Temperatura):C(Presion) + C(Temperatura):C(Tiempo) + \
            C(Presion):C(Tiempo) + C(Temperatura):C(Presion):C(Tiempo)', data=df).fit()

# Tabla ANOVA completa
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# TABLA ANOVA FORMATEADA
print("\n" + "="*50)
print("TABLA ANOVA COMPLETADA")
print("="*50)

# Crear tabla formateada similar a la del problema
tabla_formateada = pd.DataFrame({
    'Fuente': ['A (Temperatura)', 'B (Presión)', 'C (Tiempo)',
               'AB', 'AC', 'BC', 'ABC', 'ERROR', 'TOTAL'],
    'Suma_Cuadrados': [0]*9,
    'GI': [0]*9,
    'Cuadrado_Medio': [0]*9,
    'F_calculada': [0]*9
})

# Llenar la tabla con los resultados
fuentes_anova = {
    'A (Temperatura)': 'C(Temperatura)',
    'B (Presión)': 'C(Presion)',
    'C (Tiempo)': 'C(Tiempo)',
    'AB': 'C(Temperatura):C(Presion)',
    'AC': 'C(Temperatura):C(Tiempo)',
    'BC': 'C(Presion):C(Tiempo)',
    'ABC': 'C(Temperatura):C(Presion):C(Tiempo)'
}

for i, (fuente, col_anova) in enumerate(fuentes_anova.items()):
    if col_anova in anova_table.index:
        tabla_formateada.loc[i, 'Suma_Cuadrados'] = anova_table.loc[col_anova, 'sum_sq']
        tabla_formateada.loc[i, 'GI'] = anova_table.loc[col_anova, 'df']
        tabla_formateada.loc[i, 'Cuadrado_Medio'] = anova_table.loc[col_anova, 'sum_sq'] / anova_table.loc[col_anova, 'df']
        tabla_formateada.loc[i, 'F_calculada'] = anova_table.loc[col_anova, 'F']

# Error y Total
tabla_formateada.loc[7, 'Suma_Cuadrados'] = anova_table.loc['Residual', 'sum_sq']
tabla_formateada.loc[7, 'GI'] = anova_table.loc['Residual', 'df']
tabla_formateada.loc[7, 'Cuadrado_Medio'] = anova_table.loc['Residual', 'sum_sq'] / anova_table.loc['Residual', 'df']

tabla_formateada.loc[8, 'Suma_Cuadrados'] = anova_table['sum_sq'].sum()
tabla_formateada.loc[8, 'GI'] = anova_table['df'].sum()

# Mostrar tabla formateada
print(tabla_formateada.round(4))

# ANÁLISIS DE SIGNIFICANCIA (α = 0.05)
print("\n" + "="*50)
print("ANÁLISIS DE SIGNIFICANCIA (α = 0.05)")
print("="*50)

from scipy.stats import f

alpha = 0.05
for i, row in tabla_formateada.iterrows():
    if row['Fuente'] not in ['ERROR', 'TOTAL'] and row['GI'] > 0:
        f_critico = f.ppf(1 - alpha, row['GI'], tabla_formateada.loc[7, 'GI'])
        significativo = "SÍ" if row['F_calculada'] > f_critico else "NO"
        print(f"{row['Fuente']}: F = {row['F_calculada']:.4f}, F crítico = {f_critico:.4f} → Significativo: {significativo}")

# GRÁFICOS DE INTERACCIÓN
print("\n" + "="*50)
print("GRÁFICOS DE INTERACCIÓN")
print("="*50)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Interacción Temperatura x Presión
temp_pres = df.groupby(['Temperatura', 'Presion'])['Rendimiento'].mean().unstack()
temp_pres.plot(ax=axes[0,0], marker='o')
axes[0,0].set_title('Interacción Temperatura x Presión')
axes[0,0].set_ylabel('Rendimiento Promedio')
axes[0,0].grid(True, alpha=0.3)

# Interacción Temperatura x Tiempo
temp_tiempo = df.groupby(['Temperatura', 'Tiempo'])['Rendimiento'].mean().unstack()
temp_tiempo.plot(ax=axes[0,1], marker='s')
axes[0,1].set_title('Interacción Temperatura x Tiempo')
axes[0,1].set_ylabel('Rendimiento Promedio')
axes[0,1].grid(True, alpha=0.3)

# Interacción Presión x Tiempo
pres_tiempo = df.groupby(['Presion', 'Tiempo'])['Rendimiento'].mean().unstack()
pres_tiempo.plot(ax=axes[1,0], marker='^')
axes[1,0].set_title('Interacción Presión x Tiempo')
axes[1,0].set_ylabel('Rendimiento Promedio')
axes[1,0].grid(True, alpha=0.3)

# Efectos principales
medias_factores = pd.DataFrame({
    'Temperatura': df.groupby('Temperatura')['Rendimiento'].mean(),
    'Presión': df.groupby('Presion')['Rendimiento'].mean(),
    'Tiempo': df.groupby('Tiempo')['Rendimiento'].mean()
})
medias_factores.plot(ax=axes[1,1], kind='bar')
axes[1,1].set_title('Efectos Principales de los Factores')
axes[1,1].set_ylabel('Rendimiento Promedio')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# RESUMEN FINAL
print("\n" + "="*50)
print("RESUMEN Y CONCLUSIONES")
print("="*50)

print("Factores significativos al nivel α = 0.05:")
for i, row in tabla_formateada.iterrows():
    if row['Fuente'] not in ['ERROR', 'TOTAL'] and row['GI'] > 0:
        f_critico = f.ppf(1 - alpha, row['GI'], tabla_formateada.loc[7, 'GI'])
        if row['F_calculada'] > f_critico:
            print(f"  ✓ {row['Fuente']} (F = {row['F_calculada']:.4f}, p < 0.05)")

print("\nRecomendaciones:")
print("1. Identificar las condiciones óptimas basándose en los factores significativos")
print("2. Considerar las interacciones significativas en el proceso")
print("3. Validar los resultados con experimentos de confirmación")
```

## **5. Calculo LSD para el efecto Simple A**

```python
import pandas as pd
import numpy as np
from scipy.stats import t

# Calcular LSD para el efecto simple A (Temperatura)
print("CÁLCULO DE LSD PARA EL EFECTO SIMPLE A (TEMPERATURA)")
print("="*50)

# Parámetros del ANOVA
alpha = 0.05
CM_error = 1.16667  # Cuadrado medio del error de la tabla ANOVA
gl_error = 24        # Grados de libertad del error
n_replicas = 3       # Réplicas por combinación de factores

# Niveles de temperatura y sus medias
temperaturas = [100, 120, 140]
medias_temperatura = df.groupby('Temperatura')['Rendimiento'].mean()

print("Medias por temperatura:")
for temp in temperaturas:
    print(f"  Temperatura {temp}: {medias_temperatura[temp]:.4f}")

# Cálculo del LSD
t_critico = t.ppf(1 - alpha/2, gl_error)  # Valor t de dos colas
print(f"\nParámetros:")
print(f"  Cuadrado Medio del Error (CME): {CM_error}")
print(f"  Grados de libertad del error: {gl_error}")
print(f"  Número de réplicas por tratamiento: {n_replicas}")
print(f"  t crítico (α={alpha}, gl={gl_error}): {t_critico:.4f}")

# Para efecto simple A, el número de observaciones por nivel de temperatura
n_obs_por_temp = len(pressures) * len(times) * n_replicas
print(f"  Observaciones por nivel de temperatura: {n_obs_por_temp}")

# Fórmula del LSD para efecto principal
LSD_principal = t_critico * np.sqrt(2 * CM_error / n_obs_por_temp)
print(f"\nLSD para efecto PRINCIPAL A:")
print(f"  LSD = t * √(2 * CME / n)")
print(f"  LSD = {t_critico:.4f} * √(2 * {CM_error} / {n_obs_por_temp})")
print(f"  LSD = {LSD_principal:.4f}")

# COMPARACIONES MÚLTIPLES ENTRE TEMPERATURAS
print("\n" + "="*50)
print("COMPARACIONES MÚLTIPLES ENTRE TEMPERATURAS")
print("="*50)

comparaciones = []
for i in range(len(temperaturas)):
    for j in range(i + 1, len(temperaturas)):
        temp1 = temperaturas[i]
        temp2 = temperaturas[j]
        diff = abs(medias_temperatura[temp1] - medias_temperatura[temp2])
        significativa = "SÍ" if diff > LSD_principal else "NO"

        comparaciones.append({
            'Comparación': f'{temp1} vs {temp2}',
            'Diferencia': diff,
            'LSD': LSD_principal,
            'Significativa': significativa
        })

        print(f"{temp1} vs {temp2}:")
        print(f"  Diferencia: {diff:.4f}")
        print(f"  LSD: {LSD_principal:.4f}")
        print(f"  ¿Significativa? {significativa}")
        print()

# Crear tabla de resultados
df_comparaciones = pd.DataFrame(comparaciones)
print("Resumen de comparaciones:")
print(df_comparaciones)

# LSD PARA EFECTO SIMPLE (A dentro de niveles de otros factores)
print("\n" + "="*50)
print("LSD PARA EFECTOS SIMPLES (A dentro de B y C)")
print("="*50)

# Para efecto simple, el número de observaciones es menor
n_obs_efecto_simple = n_replicas  # Solo las réplicas dentro de cada combinación
LSD_simple = t_critico * np.sqrt(2 * CM_error / n_obs_efecto_simple)

print(f"LSD para efecto SIMPLE A (dentro de niveles fijos de B y C):")
print(f"  LSD = t * √(2 * CME / n_simple)")
print(f"  LSD = {t_critico:.4f} * √(2 * {CM_error} / {n_obs_efecto_simple})")
print(f"  LSD = {LSD_simple:.4f}")

# COMPARACIONES PARA EFECTO SIMPLE (ejemplo dentro de B=400, C=30)
print("\nEjemplo: Comparaciones de temperatura dentro de Presión=400, Tiempo=30")
condicion = (df['Presion'] == 400) & (df['Tiempo'] == 30)
medias_simple = df[condicion].groupby('Temperatura')['Rendimiento'].mean()

print("Medias dentro de B=400, C=30:")
for temp in temperaturas:
    print(f"  Temperatura {temp}: {medias_simple[temp]:.4f}")

comparaciones_simple = []
for i in range(len(temperaturas)):
    for j in range(i + 1, len(temperaturas)):
        temp1 = temperaturas[i]
        temp2 = temperaturas[j]
        diff = abs(medias_simple[temp1] - medias_simple[temp2])
        significativa = "SÍ" if diff > LSD_simple else "NO"

        comparaciones_simple.append({
            'Condición': 'B=400, C=30',
            'Comparación': f'{temp1} vs {temp2}',
            'Diferencia': diff,
            'LSD_simple': LSD_simple,
            'Significativa': significativa
        })

        print(f"  {temp1} vs {temp2}: Diferencia = {diff:.4f}, LSD = {LSD_simple:.4f} → {significativa}")

# VISUALIZACIÓN GRÁFICA
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Gráfico 1: Medias principales con LSD
plt.subplot(1, 2, 1)
bars = plt.bar([str(t) for t in temperaturas], medias_temperatura.values,
               color=['skyblue', 'lightgreen', 'lightcoral'], alpha=0.7)
plt.axhline(y=medias_temperatura.mean(), color='red', linestyle='--',
           label=f'Media general: {medias_temperatura.mean():.2f}')
plt.errorbar([str(t) for t in temperaturas], medias_temperatura.values,
             yerr=LSD_principal/2, fmt='none', color='black', capsize=5, label='LSD/2')

plt.xlabel('Temperatura')
plt.ylabel('Rendimiento Promedio')
plt.title('Medias de Temperatura con Intervalo LSD\n(Efecto Principal)')
plt.legend()
plt.grid(True, alpha=0.3)

# Gráfico 2: Comparaciones
plt.subplot(1, 2, 2)
diferencias = [comp['Diferencia'] for comp in comparaciones]
nombres_comparaciones = [comp['Comparación'] for comp in comparaciones]
colors = ['green' if diff > LSD_principal else 'red' for diff in diferencias]

bars = plt.bar(nombres_comparaciones, diferencias, color=colors, alpha=0.7)
plt.axhline(y=LSD_principal, color='blue', linestyle='--',
           label=f'LSD = {LSD_principal:.3f}')

plt.xlabel('Comparaciones')
plt.ylabel('Diferencia Absoluta')
plt.title('Diferencias entre Temperaturas vs LSD')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# RESUMEN FINAL
print("\n" + "="*50)
print("RESUMEN FINAL - PRUEBA LSD")
print("="*50)

print(f"LSD para efecto PRINCIPAL A: {LSD_principal:.4f}")
print(f"LSD para efecto SIMPLE A: {LSD_simple:.4f}")

print("\nInterpretación:")
print("• Si la diferencia entre dos medias de temperatura es MAYOR que el LSD,")
print("  la diferencia es estadísticamente significativa (α = 0.05)")
print("• Para el efecto principal, se comparan las medias globales de cada temperatura")
print("• Para efectos simples, se comparan medias dentro de condiciones específicas")

print("\nRecomendación:")
print("Use LSD_principal para comparaciones generales entre temperaturas")
print("Use LSD_simple para comparaciones dentro de condiciones específicas de presión y tiempo")
```

## **6. Considerando el factor C como aleatorio**

Si consideramos el **Factor C (Tiempo)** como un factor **aleatorio** en el diseño factorial 3x2x2, mientras que los **Factores A (Temperatura)** y **B (Presión)** se mantienen como **fijos**, el modelo se convierte en un **modelo de efectos mixtos**. Esto cambia la forma en que se calculan las pruebas F para determinar la significancia de los efectos.

En un modelo de efectos mixtos, las pruebas F se construyen utilizando cuadrados medios esperados. Basándonos en la tabla ANOVA proporcionada anteriormente y la estructura del diseño (A con 3 niveles, B con 2, C con 2, con réplicas), los denominadores para las pruebas F cambian de la siguiente manera si C es aleatorio:

*   **Efecto Principal A (Temperatura - Fijo):** La prueba F para el factor fijo A se realiza típicamente contra el cuadrado medio de la interacción de A con el factor aleatorio C (CM_AC), si esta interacción existe en el modelo y no es cero. Si la interacción AC no es significativa o se asume cero, se usaría el CM del error.
    *   $F_A = \frac{CM_A}{CM_{AC}}$ (Si AC es significativo o se asume que no es cero)
    *   $F_A = \frac{CM_A}{CM_{Error}}$ (Si AC no es significativo o se asume cero)

*   **Interacción AC (Fijo x Aleatorio):** La prueba F para la interacción entre un factor fijo y uno aleatorio se realiza típicamente contra el cuadrado medio del error (CM_Error).

    *   $F_{BC} = \frac{CM_{BC}}{CM_{Error}}$

*   **Interacción ABC (Fijo x Fijo x Aleatorio):** La prueba F para la interacción de orden superior que involucra el factor aleatorio se realiza contra el cuadrado medio del error (CM_Error).

## **7. Anova con el factor C aleatorio**

```python
import pandas as pd
from scipy.stats import f

# Extract values from the provided ANOVA table (cell KCUOAK9vWhO2)
# Note: These values are assumed to be correct from the user's previous analysis.
# We are using these pre-calculated values to demonstrate the F-tests for a mixed model.

# Sums of Squares (SS) and Degrees of Freedom (GI/df) from the provided table
SS = {
    'A': 22.3889, 'B': 10.0278, 'C': 283.3611,
    'AB': 2.72222, 'AC': 3.72222, 'BC': 1.36111,
    'ABC': 0.055556, 'ERROR': 28.0000
}

GI = {
    'A': 2, 'B': 1, 'C': 1,
    'AB': 2, 'AC': 2, 'BC': 1,
    'ABC': 2, 'ERROR': 24
}

# Calculate Mean Squares (CM)
CM = {source: SS[source] / GI[source] if GI[source] > 0 else 0 for source in SS}

print("Mean Squares from provided ANOVA table:")
for source, value in CM.items():
    print(f"  CM_{source}: {value:.4f}")

# Determine the appropriate denominators for F-tests in a mixed model (C is Random)
# Based on expected mean squares:
# F_A tested against CM_AC
# F_B tested against CM_BC
# F_C tested against CM_ABC
# F_AB tested against CM_ABC
# F_AC tested against CM_Error
# F_BC tested against CM_Error
# F_ABC tested against CM_Error

# Calculate the F-statistics using the mixed model denominators
F_calculated = {}
F_denominators_GI = {} # To store denominator degrees of freedom for F-tests

# F for A
denominator_A = CM['AC'] if GI['AC'] > 0 else CM['ERROR'] # Use CM_AC if GI_AC > 0, otherwise CM_Error
F_calculated['A'] = CM['A'] / denominator_A if denominator_A > 0 else float('inf')
F_denominators_GI['A'] = GI['AC'] if GI['AC'] > 0 else GI['ERROR']

# F for B
denominator_B = CM['BC'] if GI['BC'] > 0 else CM['ERROR'] # Use CM_BC if GI_BC > 0, otherwise CM_Error
F_calculated['B'] = CM['B'] / denominator_B if denominator_B > 0 else float('inf')
F_denominators_GI['B'] = GI['BC'] if GI['BC'] > 0 else GI['ERROR']

# F for C (Random)
denominator_C = CM['ABC'] if GI['ABC'] > 0 else CM['ERROR'] # Use CM_ABC if GI_ABC > 0, otherwise CM_Error
F_calculated['C'] = CM['C'] / denominator_C if denominator_C > 0 else float('inf')
F_denominators_GI['C'] = GI['ABC'] if GI['ABC'] > 0 else GI['ERROR']

# F for AB
denominator_AB = CM['ABC'] if GI['ABC'] > 0 else CM['ERROR'] # Use CM_ABC if GI_ABC > 0, otherwise CM_Error
F_calculated['AB'] = CM['AB'] / denominator_AB if denominator_AB > 0 else float('inf')
F_denominators_GI['AB'] = GI['ABC'] if GI['ABC'] > 0 else GI['ERROR']

# F for AC
denominator_AC = CM['ERROR'] if GI['ERROR'] > 0 else float('inf') # Test against Error
F_calculated['AC'] = CM['AC'] / denominator_AC if denominator_AC > 0 else float('inf')
F_denominators_GI['AC'] = GI['ERROR']

# F for BC
denominator_BC = CM['ERROR'] if GI['ERROR'] > 0 else float('inf') # Test against Error
F_calculated['BC'] = CM['BC'] / denominator_BC if denominator_BC > 0 else float('inf')
F_denominators_GI['BC'] = GI['ERROR']

# F for ABC
denominator_ABC = CM['ERROR'] if GI['ERROR'] > 0 else float('inf') # Test against Error
F_calculated['ABC'] = CM['ABC'] / denominator_ABC if denominator_ABC > 0 else float('inf')
F_denominators_GI['ABC'] = GI['ERROR']

print("\nCalculated F-statistics (C is Random):")
for source, value in F_calculated.items():
    print(f"  F_{source}: {value:.4f}")

# Determine significance by comparing F_calculated with critical F values (α = 0.05)
alpha = 0.05
significance = {}

print("\nSignificance Analysis (α = 0.05):")
for source, F_value in F_calculated.items():
    df_num = GI[source]
    df_den = F_denominators_GI[source]

    if df_num > 0 and df_den > 0:
        try:
            f_critico = f.ppf(1 - alpha, df_num, df_den)
            is_significant = F_value > f_critico
            significance[source] = "SÍ" if is_significant else "NO"
            print(f"  {source}: F_calculated = {F_value:.4f}, F_critical({df_num},{df_den}) = {f_critico:.4f} -> Significant: {significance[source]}")
        except ValueError as e:
            print(f"  Could not calculate F_critical for {source} ({df_num},{df_den}): {e}")
            significance[source] = "Cannot calculate"
    else:
        significance[source] = "N/A (GI = 0)"
        print(f"  {source}: N/A (GI = 0)")

print("\nSummary of Significant Effects (C is Random, α = 0.05):")
significant_effects = [source for source, sig in significance.items() if sig == "SÍ"]
if significant_effects:
    print("The following effects are statistically significant:")
    for effect in significant_effects:
        print(f"  - {effect}")
else:
    print("No effects are statistically significant at α = 0.05 with C as a random factor based on this analysis.")

print("\nNote: This analysis uses the Mean Squares from the provided ANOVA table.")
print("A full mixed-effects model analysis would typically be performed using specialized software or libraries.")
print("The F-tests for A and B against interactions AC and BC have very low denominator degrees of freedom,")
print("which results in high critical F values and reduced power to detect significance for A and B using this test.")
print("If AC or BC interactions were found to be non-significant, alternative approximate F-tests or 'pooling' of error terms might be considered in practice, but testing against the interaction with the random factor is the standard approach if that interaction is non-zero.")
```
